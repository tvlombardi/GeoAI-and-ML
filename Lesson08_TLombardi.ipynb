#### 2. Implement the haversine function to calculate the distance between two lon/lat coordinates and vectorize it using numpy

```python
# Adapted formula from comments in https://gist.github.com/rochacbruno/2883505
def distance_vectorize(s_lat, s_lng, e_lat, e_lng, df):
    
    # approximate radius of earth in km
    R = 6371.0
    
    s_lat = s_lat*np.pi/180.0                      
    s_lng = np.deg2rad(s_lng)     
    e_lat = np.deg2rad(e_lat)                       
    e_lng = np.deg2rad(e_lng)  
    
    d = np.sin((e_lat - s_lat)/2)**2 + np.cos(s_lat)*np.cos(e_lat) * np.sin((e_lng - s_lng)/2)**2
    
    df['haversine_distance_km'] = 2 * R * np.arcsin(np.sqrt(d))
```

#### 9. Identify clusters with [K-Means](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

[Sklearn.clustering](https://scikit-learn.org/stable/modules/clustering.html#k-means)
>"The k-means algorithm divides a set of **N** samples **X** into **K** disjoint clusters **C**, each described by the mean (μj) of the samples in the cluster. The means are commonly called the cluster “centroids”.

>In basic terms, the algorithm has three steps. The first step chooses the initial centroids, with the most basic method being to choose **k** samples from the dataset **X**. After initialization, K-means consists of looping between the two other steps. The first step assigns each sample to its nearest centroid. The second step creates new centroids by taking the mean value of all of the samples assigned to each previous centroid. The difference between the old and the new centroids are computed and the algorithm repeats these last two steps until this value is less than a threshold. In other words, it repeats until the centroids do not move significantly.

```python
from sklearn.cluster import KMeans
```

There are three steps to preparing the data: 
- create the coordinates stack, 
- configure the KMeans clustering parameters, and 
- create the actual clusters

##### Use K-Means to cluster the pickup locations<a id='pickup_cluster'></a>

```python
# Create the coordinates stack
coords = np.vstack(train[['pickup_latitude', 'pickup_longitude']].values)
city_long_border = (-74.03, -73.75)
city_lat_border = (40.63, 40.85)
```

```python
# Configure the KMeans clustering parameters
km = KMeans(n_clusters=100)
km.fit(coords)
```

```python
# Create the clusters and give each row a pickup_cluster ID
train.loc[:, 'pickup_cluster'] = km.predict(train[['pickup_latitude', 'pickup_longitude']])
```

```python
# Plot the clusters
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
centers = km.cluster_centers_
ax.scatter(train.pickup_longitude.values[:500000], 
           train.pickup_latitude.values[:500000],
           s=10, 
           lw=0,
           c=train.pickup_cluster[:500000].values, 
           cmap='autumn', 
           alpha=0.2)
ax.set_xlim(city_long_border)
ax.set_ylim(city_lat_border)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```

##### Use K-Means to cluster the pickup hours<a id='hours_cluster'></a>

```python
# Create the hour and day of week stack
hod = np.vstack(train[['Hour', 'DayofWeek']].values)
```

```python
# Configure the KMeans clustering parameters
km = KMeans(n_clusters=10)
km.fit(hod)
```

```python
# Create the clusters and give each row a hod_cluster ID
train.loc[:, 'hod_cluster'] = km.predict(train[['Hour', 
                                                'DayofWeek']])
```

```python
# Plot the clusters
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
centers = km.cluster_centers_
ax.scatter(train.pickup_longitude.values[:500000], 
           train.pickup_latitude.values[:500000],
           s=10, 
           lw=0,
           c=train.hod_cluster[:500000].values, 
           cmap='tab10', 
           alpha=0.2)
ax.set_xlim(city_long_border)
ax.set_ylim(city_lat_border)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
fig.set_figheight(15)
fig.set_figwidth(15)
plt.show()
```
